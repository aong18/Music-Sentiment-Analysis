```
import pandas as pd
artists_data = pd.read_csv("archive/artists-data.csv")
lyrics_data = pd.read_csv("archive/lyrics-data.csv")
song_totals = artists_data.sort_values(by=["Songs"], ascending = False)
# song_totals.head(10)
```

```
lyrics_data_eng = lyrics_data[lyrics_data['Idiom'] == 'ENGLISH']
merged_artists = pd.merge(artists_data, lyrics_data_eng, left_on="Link", right_on="ALink").drop_duplicates("Artist")
merged_artists
```

```
merged_artists.Genre.unique()
```

```
popularity_list = merged_artists.sort_values(by=["Popularity"], ascending = False)
most_popular = popularity_list.drop_duplicates("Artist")
ten_most_popular = most_popular.iloc[:10]
ten_most_popular
```

```
kanye_info = artists_data[artists_data['Artist'] == 'Kanye West']
```

```
kanye_songs = lyrics_data[lyrics_data['ALink'] == '/kanye-West/']
kanye_songs
```

```
popular_hip_hop = most_popular[most_popular["Genre"] == "Hip Hop"]
most_popular_hip_hop = popular_hip_hop.iloc[:10]
most_popular_hip_hop
```

```
popular_pop = most_popular[most_popular["Genre"] == "Pop"]
most_popular_pop = popular_pop.iloc[:10]
most_popular_pop
```

```
popular_rock = most_popular[most_popular["Genre"] == "Rock"]
most_popular_rock = popular_rock.iloc[:10]
most_popular_rock
```

```
import nltk
import math
import re
import string
from itertools import product

import nltk.data
from nltk.util import pairwise
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')

analyzer = SentimentIntensityAnalyzer()
```

```
kanye_lyrics = kanye_songs['Lyric']
kanye_songs['compound'] = kanye_lyrics.apply(lambda x:analyzer.polarity_scores(x)['compound'])
kanye_songs['compound'].describe()
```

```
ariana_songs = lyrics_data[lyrics_data['ALink'] == '/ariana-grande/']
ariana_lyrics = ariana_songs['Lyric']
ariana_songs['compound'] = ariana_lyrics.apply(lambda x:analyzer.polarity_scores(x)['compound'])
#ariana_songs['compound'].describe()
```

```
queen_songs = lyrics_data[lyrics_data['ALink'] == '/queen/']
queen_lyrics = queen_songs['Lyric']
queen_songs['compound'] = queen_lyrics.apply(lambda x:analyzer.polarity_scores(x)['compound'])
#ariana_songs['compound'].describe()
```

```
import matplotlib.pyplot as plt
import seaborn as sns
#plt.figure(figsize = (9, 7))
#sns.distplot(kanye_songs['compound'], label = 'Kanye')
#sns.distplot(adele_songs['compound'], label = 'Adele')
#sns.distplot(ariana_songs['compound'], label = 'Ariana')
#plt.xlabel('polarity')
#plt.ylabel ('Density')
#plt.title("Distributions of Song Lyrics Polarity")
#plt.legend()
```

```
df = pd.DataFrame(columns=('artist', 'pos', 'neu', 'neg'))

num_positive = 0
num_negative = 0
num_neutral = 0

for compound in kanye_songs['compound']:
    if compound >= 0.5:
        num_positive += 1
    elif compound > -0.5 and compound < 0.5:
        num_neutral += 1
    else:
        num_negative += 1

num_total = num_negative + num_neutral + num_positive
percent_negative = (num_negative/float(num_total))*100
percent_neutral = (num_neutral/float(num_total))*100
percent_positive = (num_positive/float(num_total))*100
df.loc[0] = ('Kanye West ', percent_positive, percent_neutral, percent_negative)


for compound in queen_songs['compound']:
    if compound >= 0.5:
        num_positive += 1
    elif compound > -0.5 and compound < 0.5:
        num_neutral += 1
    else:
        num_negative += 1

num_total = num_negative + num_neutral + num_positive
percent_negative = (num_negative/float(num_total))*100
percent_neutral = (num_neutral/float(num_total))*100
percent_positive = (num_positive/float(num_total))*100
df.loc[1] = ('Queen', percent_positive, percent_neutral, percent_negative)


for compound in ariana_songs['compound']:
    if compound >= 0.5:
        num_positive += 1
    elif compound > -0.5 and compound < 0.5:
        num_neutral += 1
    else:
        num_negative += 1

num_total = num_negative + num_neutral + num_positive
percent_negative = (num_negative/float(num_total))*100
percent_neutral = (num_neutral/float(num_total))*100
percent_positive = (num_positive/float(num_total))*100
df.loc[2] = ('Ariana Grande', percent_positive, percent_neutral, percent_negative)

                 
df.plot.bar(x='artist', stacked=True)
plt.xlabel("Artist")
plt.show()
```

```
test_df = pd.DataFrame(columns=('artist', 'pos', 'neu', 'neg'))

counter = 0
for artist in ten_most_popular["Link"].to_list():
    num_positive = 0
    num_negative = 0
    num_neutral = 0
    artist_songs = lyrics_data_eng[lyrics_data_eng['ALink'] == artist]
    artist_lyrics = artist_songs['Lyric']
    artist_songs['compound'] = artist_lyrics.apply(lambda x:analyzer.polarity_scores(x)['compound'])

    for compound in artist_songs['compound']:
        if compound >= 0.5:
            num_positive += 1
        elif compound > -0.5 and compound < 0.5:
            num_neutral += 1
        else:
            num_negative += 1

    num_total = num_negative + num_neutral + num_positive
    percent_negative = (num_negative/float(num_total))*100
    percent_neutral = (num_neutral/float(num_total))*100
    percent_positive = (num_positive/float(num_total))*100
    artist = artist[1:-1]
    test_df.loc[counter] = (artist, percent_positive, percent_neutral, percent_negative)
    counter += 1

test_df.plot.bar(x='artist', stacked=True)
plt.xlabel('Most Popular Artists')
plt.show()
```
